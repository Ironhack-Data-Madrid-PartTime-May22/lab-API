```python
# Importo las librerias que vamos necesitar para el lab.
import pandas as pd 
pd.options.display.max_columns = None 
import requests  

# Funcion para hacer la llamada a la API
def get_api(url):
    res = requests.get(url) 
    if res.status_code == 200:  
        return res.json() 
    else:  
        return f"Error {res.status_code}"

url = "https://db.ygoprodeck.com/api/v7/cardinfo.php"
res_api = get_api(url)

# Revisamos la informacion y podemos ver que solo hay una lista de diccionarios que tiene como key 'data'
res_api.keys()

# Dentro de ella esta toda la informacion de las cartas, entonces gurdamos esta lista de diccionarios dentro de una variable
diccionario_limpio = res_api["data"]

# Revisamos las keys que estan dentro
res_api["data"][0].keys()
res_api['data'][1000].keys() # Probamos para poder ver todas la diferentes keys que hay, ya que en la informacion dice que no todos tienen las mismas


# Descartamos la informacion que no nos interesa como: 'card_sets', 'card_images'y 'card_prices' 
# y hacemos un diccionario vacio con las keys que si queremos
info = {
    "id_": [],
    "name": [],
    "type": [],
    "frametype": [],
    "desc": [],
    "race": [],
    "archetype": [],
    "atk": [],
    "def": [],
    "level": [],
    "attribute": [],
    "url_": []
}

# Iteramos por todos los valores que tenemos en la key "data" guardado previamente en la variable diccionario_limpio
# Uso el metodo 'get' para evitar que se produzcan excepciones si alguna key no estÃ¡ presente en todos los elementos del diccionario,
# al usar este metodo puedo introducir None en su lugar
for i in diccionario_limpio:
    info["id_"].append(i.get("id", None))
    info["name"].append(i.get("name", None))
    info["type"].append(i.get("type", None))
    info["frametype"].append(i.get("frameType", None))
    info["desc"].append(i.get("desc", None))
    info["race"].append(i.get("race", None))
    info["archetype"].append(i.get("archetype", None))
    info["atk"].append(i.get("atk", None))
    info["def"].append(i.get("def", None))
    info["level"].append(i.get("level", None))
    info["attribute"].append(i.get("attribute", None))

# Quiero que se pueda ver la informacion mas detallada y la imagen de la carta, asi que busque una pagina que cambiando el nombre me saque la url de cada carta
# Itero para poder pasar los espacios a barras bajas, y los meto dentro del diccionario vacio
name_search = {"name_search": []}

for e in info["name"]:
    n = e.replace(" ", "_")  # Reemplazar espacios por barra baja en cada nombre
    name_search["name_search"].append(n)  # Agregar el nombre modificado a la lista

# Hago un bucle para poder meter los urls dentro de info para que se vea en mi df
for i in name_search["name_search"]:
    url_search = f"https://yugioh.fandom.com/wiki/{i}" # Hacemos cada url diferente con "name_search"
    info["url_"].append(url_search) #Agregamos cada url dentro de "info['url_']" 

# Para concluir, hago un DataFrame con mi diccionario "info" que tiene toda la informacion que con anterioridad he limpiado.
df = pd.DataFrame(info)

# Pongo como indice la columna "id_"
df.set_index('id_', inplace=True)

df


    




## Link de donde he sacado y revisado la informacion: https://ygoprodeck.com/api-guide/
```
