```python

import pandas as pd
pd.options.display.max_columns = None 


import requests


import os 
from dotenv import load_dotenv 

#importo todo lo que tengo que importar
```

```python
def get_api(url):
    res = requests.get(url) 
    if res.status_code == 200: 
        return res.json() 
    else: 
        return f"Error {res.status_code}"
    
#definimos una funci√≥n que llama a la API. Si la llamada es correcta, nos da los datos en formato JSON. 
# Si la llamada no es correcta, nos levanta un error
```

```python
url = 'https://api.disneyapi.dev/character'

res_api = get_api(url)

res_api

#cogemos la URL seleccionada (personajes de Disney), y hacemos la llamada a la API y mostramos los datos
```

```python
data = res_api['data'] #de las keys de res_api, a m√≠ me interesa la info de 'data', por lo que la guardo en una nueva variable

informacion = {'id': [], 'nombre': [], 'source_URL': []} #creo un diccionario vac√≠o con los datos que quiero extraer de los datos cogidos de la API


for i in data: 
    informacion['id'].append(i['_id'])
    informacion['nombre'].append(i['name'])
    informacion['source_URL'].append(i['sourceUrl'])  
    #creo un bucle que itere por todos los diccionarios de la lista devuelta por la API, y voy apendeando los datos que me interesan (id, nombre y sourceURL)

df = pd.DataFrame(informacion) #una vez creado el diccionario con los datos, creo un dataframe

df

```
